## 1. Read files

```python
import numpy

file = open("data/global-electricity-consumption.csv")
years = file.readline().strip().split(",")[1::]

countries = numpy.genfromtxt("data/global-electricity-consumption.csv", delimiter=",", skip_header=True, dtype=str, usecols=[0])
consumption = numpy.genfromtxt("data/global-electricity-consumption.csv", delimiter=",", skip_header=True, usecols=[i for i in range(1, len(years) + 1)])
generation = numpy.genfromtxt("data/global-electricity-generation.csv", delimiter=",", skip_header=True, usecols=[i for i in range(1, len(years) + 1)])
```

## 2. Consumption and generation for the last 5 years by country

```python
avgConsumption = numpy.mean(consumption[:, -5:], axis=1)
avgGeneration = numpy.mean(generation[:, -5:], axis=1)
for i in range(len(countries)):
    print(f"{countries[i]}: consumption = {avgConsumption[i]:.2f} kW*h, generation = {avgGeneration[i]:.2f} kW*h")
```

## 3. Questions

### 3.1 Sum of consumption by year

```python
consumptionSum = numpy.nansum(consumption, axis=0)
for i in range(len(years)):
    print(f"{years[i]}: {consumptionSum[i]:.2f} kW*h")
```

### 3.2. Max yearly generation by country

```python
maxGenerationIndices = numpy.nanargmax(generation, axis=1)
for i in range(len(countries)):
    print(f"{countries[i]}: {generation[i][maxGenerationIndices[i]]:.2f} TW*h in {years[maxGenerationIndices[i]]}")
```

### 3.3. Contries producing more than 500 TW*h for the last 5 years

```python
print(", ".join(countries[avgGeneration > 500].tolist()))
```

### 3.4. Top 10% of consumption by country

```python
percentile = numpy.percentile(avgConsumption, 90)
top10 = countries[avgConsumption > percentile]
print(", ".join(top10.tolist()))
```

### 3.5. Coutires which increased their generation more than 10x between 1992 and 2021

```python
i1992 = years.index("1992")
i2021 = years.index("2021")

for country in countries[generation[:, i1992] * 10 < generation[:, i2021]]:
    print(country)
```

### 3.6 Coutries which consumed more than 100 TW*h from 1992 to 2021 but generated less than that 

```python
consumptionSum = numpy.nansum(consumption, axis=1)
generationSum = numpy.nansum(generation, axis=1)
bigConsumptionMask = consumptionSum > 100
lowGenerationMask = consumptionSum > generationSum 
resultMask = bigConsumptionMask
for index, item in enumerate(lowGenerationMask):
    resultMask[index] = resultMask[index] and item

for country in countries[resultMask]:
    print(country)
```

### 3.7 Country with max consumpotion in 2020

```python
i2020 = years.index("2020")

max = numpy.nanmax(consumption[:, i2020], axis=0)
print(countries[consumption[:, i2020] == max][0] + ":", max, "TW*h")
```

# Task 2

```python
from scipy import linalg
import matplotlib.pyplot as plt

data2 = numpy.genfromtxt("data/data2.csv", delimiter=";", dtype=float, skip_header=True)
dataset_len = len(data2)
indices_2 = [0, dataset_len // 2, dataset_len - 1]
indices_3 = [0, dataset_len // 3, dataset_len // 3 * 2, dataset_len - 1]
x_and_res_2 = data2[indices_2]
x_and_res_3 = data2[indices_3]

koeffs_2 = []
b_2 = []
for [x, res] in x_and_res_2:
    koeffs_2.append([x**2, x, 1.0])
    b_2.append(res)

koeffs_2 = numpy.array(koeffs_2)
b_2 = numpy.array(b_2)

koeffs_3 = []
b_3 = []
for [x, res] in x_and_res_3:
    koeffs_3.append([x**3, x**2, x, 1.0])
    b_3.append(res)

koeffs_3 = numpy.array(koeffs_3)
b_3 = numpy.array(b_3)

solution_2 = linalg.solve(koeffs_2, b_2)
solution_3 = linalg.solve(koeffs_3, b_3)
```

### График для полинома второй степени

```python
x_fit = numpy.linspace(0, 5, 200)
[a2, a1, a0] = solution_2
y_fit = a2 * x_fit**2 + a1 * x_fit + a0

plt.figure(figsize=(10, 6))
plt.scatter(data2[:, 0], data2[:, 1], color='red', label='Данные из файла', zorder=5)
plt.plot(x_fit, y_fit, color='blue', linewidth=2, label=f'Полином 2-й степени')
plt.xlabel('Скидка')
plt.ylabel('Прибыль')
plt.title('Зависимость прибыли от скидки')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
RSS_2 = 0
for [x, y] in data2:
    RSS_2 += (y - (solution_2[0] * x**2 + solution_2[1] * x + solution_2[2]))**2
print('RSS =', RSS_2)
```

### График для полинома третьей степени

```python
x_fit = numpy.linspace(0, 8, 200)
[a3, a2, a1, a0] = solution_3
y_fit = a3 * x_fit**3 + a2 * x_fit**2 + a1 * x_fit + a0

plt.figure(figsize=(10, 6))
plt.scatter(data2[:, 0], data2[:, 1], color='red', label='Данные из файла', zorder=5)
plt.plot(x_fit, y_fit, color='blue', linewidth=2, label=f'Полином 3-й степени')
plt.xlabel('Скидка')
plt.ylabel('Прибыль')
plt.title('Зависимость прибыли от скидки')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
RSS_3 = 0
for [x, y] in data2:
    RSS_3 += (y - (solution_3[0] * x**3 + solution_3[1] * x**2 + solution_3[2] * x + solution_3[3]))**2
print('RSS =', RSS_3)
```

RSS значительно меньше у полинома третьей степени. Используем его для вычисления прибыли при скидке 6 и 8 процентов.

```python
[a3, a2, a1, a0] = solution_3

def revenue(discount):
    return a3 * discount**3 + a2 * discount**2 + a1 * discount + a0

print(f'Для скидки 6% прибыль составит {revenue(6)}')
print(f'Для скидки 8% прибыль составит {revenue(8)}')
```
